```python
# In the last excerise I created a dataframe which contains the postal codes and their corresponding boroughs and neighborhoods
# This was saved as postcode_canada.csv
# Let's get the postcode dataframe from the csv saved in last excercise 
df_postcode = pd.read_csv('postcode_canada')
#Let's check the first five rows
df_postcode.head()
```

```python
# df_lat_lang and df_postcode
print('Total number of rows in df_lat_lang dataframe are: ',df_lat_lang.shape[0])
print('Total number of unique elements in "Postal Code" column in df_lat_lang dataframe are: ',len(df_lat_lang['Postal Code'].unique()))
print('Total number of rows in df_postcode dataframe are: ',df_postcode.shape[0])
print('Total number of unique elements in "Postal Code" column in df_postcode dataframe are: ',len(df_postcode['Postal Code'].unique()))
```

```python
df = pd.concat([df_lat_lang.set_index('Postal Code'),df_postcode.set_index('Postal Code')], axis=1, join='inner')
df.head()
```

```python
df = df.reset_index()
df.head()
```

```python
# Now we have one dataframe with all the information needed for the visualization of clusters
# Next we will check how many boroughs contain the word 'toronto' in it
df.Borough.unique()
#There are four Boroughs with word toronto in it
```

```python
# We seek to get all the Boroughs which have 'Toronto' in it
```

```python
# Since there are only 4 out of 8 Boroughs that have 'Toronto' in it, it is easy to copy paste and call them
# Instead we will write a code to call these without having to look at it
```

```python
# First lets make sure we change all boroughs to lower case
# Then lets split it on each space to get each word in the name of the borough
# THis will convert all the boroughs to a list of lower case words

series_borough = df.Borough.str.lower()
print(series_borough.head(20))
series_list_borough = series_borough.str.split(' ')
print(series_list_borough.head(20))
```

```python
# Then we can use apply and lambda function to find which list has toronto in it

series_toronto = series_list_borough.apply(lambda x: 'toronto' in x)
series_toronto.head(20)
```

```python
# Now we call the true values in series_toronto from df 
df_toronto = df[series_toronto]
#df_toronto = df[df.Borough.str.lower().str.split(' ').apply(lambda x: 'toronto' in x)]
df_toronto.head()
# always use the reset_index() function
```

Lets get the latitude and longitude for the city of Toronto as whole

```python
address = 'Toronto, ON, Canada'
```

```python
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
```

```python
geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
import folium
```

```python
# create map of Toronto with dataframe df using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)
# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
